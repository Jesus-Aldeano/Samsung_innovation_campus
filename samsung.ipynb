```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Cargar los datos de las dos bases de datos en DataFrames
mate= pd.read_csv('student-mat.csv', sep=';')

# Obtener una lista de las columnas que deseas mostrar
columnas_deseadas = ['school', 'sex', 'age', 'address', 'Pstatus', 'guardian', 'traveltime', 'studytime', 'failures', 'paid', 'internet', 'health', 'absences', 'G1','G2','G3']

# Filtrar el DataFrame para mostrar solo las columnas deseadas
mat= mate.loc[:, columnas_deseadas]

# Cargar los datos de las dos bases de datos en DataFrames
port= pd.read_csv('student-por.csv', sep=';')

# Obtener una lista de las columnas que deseas mostrar
columnas_deseadas = ['school', 'sex', 'age', 'address', 'Pstatus', 'guardian', 'traveltime', 'studytime', 'failures', 'paid', 'internet', 'health', 'absences', 'G1','G2','G3']

# Filtrar el DataFrame para mostrar solo las columnas deseadas
por = port.loc[:, columnas_deseadas]
```

```python

#Filtrar los datos basados en un criterio común
filtro1 = mat['school'] == 'GP'
filtro2 = por['school'] == 'GP'

datos_filtrados_mat = mat[filtro1]
datos_filtrados_por = por[filtro2]

# Realizar la combinación de los DataFrames basada en una condición específica
school_GP = pd.merge( datos_filtrados_mat, datos_filtrados_por, on='school')
school_GP
```

```python
#Filtrar los datos basados en un criterio común
filtro1 = mat['school'] == 'MS'
filtro2 = por['school'] == 'MS'

datos_filtrados_mat = mat[filtro1]
datos_filtrados_por = por[filtro2]

# Realizar la combinación de los DataFrames basada en una condición específica
school_MS = pd.merge( datos_filtrados_mat, datos_filtrados_por, on='school')
school_MS

```

```python

# Verificar si hay valores NaN en el DataFrame
hay_nan = school_MS.isna().any().any()

#  la cantidad de valores NaN en el DataFrame
cantidad_nan = school_MS.isna().sum().sum()

if hay_nan:
    print("El DataFrame contiene valores NaN.")
    print("Cantidad de valores NaN:", cantidad_nan)
else:
    print("El DataFrame no contiene valores NaN.")
```

```python

# Verificar si hay valores NaN en el DataFrame
hay_nan = school_GP.isna().any().any()

#  la cantidad de valores NaN en el DataFrame
cantidad_nan = school_GP.isna().sum().sum()

if hay_nan:
    print("El DataFrame contiene valores NaN.")
    print("Cantidad de valores NaN:", cantidad_nan)
else:
    print("El DataFrame no contiene valores NaN.")
```

```python
#la columna de interes
columna = school_GP['sex_x']

# Contar la frecuencia de cada valor en la columna
conteo = columna.value_counts()

# opciones para la gráfica de pastel
explode = [0.1, 0]  # Separación de sectores
labels = conteo.index  # Etiquetas de los sectores
color = ['tab:blue', 'tab:red']

# la gráfica de pastel
fig, ax = plt.subplots()
ax.pie(conteo, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True,colors=color, startangle=90)

# título
ax.set_title('Distribución por género de la escuela Gabriel Pereira ')
plt.show()
```

```python
columna = school_MS['sex_x']
conteo = columna.value_counts()
explode = [0.1 ,0]
labels = conteo.index
fig, ax = plt.subplots()
ax.pie(conteo, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=60)
ax.set_title('Distribución por género  de la escuela Mousinho da Silveira')
plt.show()
```

```python
columna = school_GP['age_x']
counts = columna.value_counts()
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

fig, ax = plt.subplots()
bar_container = ax.bar(counts.index, counts, color=bar_colors)
ax.axhspan(0, max(counts), facecolor='lightgray', alpha=0.5)
ax.bar_label(bar_container, fmt='{:,.0f}')
ax.set_title('cantidad de estudiantes que tienen la misma edad de la Gabriel Pereira ')
plt.show()


```

```python
promedio_edades = school_GP['age_x'].mean() * 100
print("El promedio de edades de la escuela Gabriel Pereira:",promedio_edades)
print("En porcentaje seria {:.2f}%".format(promedio_edades))
```

```python

promedio_notas_g = school_GP[['G1_y', 'G2_y', 'G3_y']].mean()
tabla_promedio = promedio_notas_g.to_frame(name='Promedio')
print("Tabla de promedio de notas:")
print(tabla_promedio)
```

```python
cmaps = ['blue', 'lime', 'pink']
fig, ax = plt.subplots()

hbars = ax.barh(promedio_notas_g.index, promedio_notas_g, align='center', color= cmaps)
ax.invert_yaxis()  # Labels leídos de arriba hacia abajo
ax.set_xlabel('Rendimiento')
ax.set_title('Grafica de promedios de la escuela  Gabriel Pereira ')

# Etiquetas con las leyendas dadas, relleno personalizado y opciones de anotación
ax.bar_label(hbars, labels=['±%.2f' % e for e in promedio_notas_g],
             padding=8, color='b', fontsize=14)
ax.set_xlim(right=16)

plt.show()
```

```python

# Calcular el valor máximo de ausencias
max_ausencias = school_GP['absences_y'].max()

# Calcular el número total de días de clase considerando el valor máximo de ausencias como el total del curso
total_dias_clase = max_ausencias + 1

# Calcular el número de días de asistencia
dias_asistencia = total_dias_clase - max_ausencias

# Calcular el porcentaje de asistencia
porcentaje_asistencia = (dias_asistencia / total_dias_clase) * 100

print("Número total de días de clase:", total_dias_clase)
print("Número de días de asistencia:", dias_asistencia)
print("Porcentaje de asistencia: {:.2f}%".format(porcentaje_asistencia))
```

```python
# Obtener el valor máximo de ausencias
max_ausencias = school_GP['absences_y'].max()

# Definir una función para aplicar a cada fila y agregar la nueva columna
def condiciones(row):
    porcentaje_asistencia = (row['absences_y'] / max_ausencias) * 100
    return round(porcentaje_asistencia, )  # Redondear a 2 decimales

# Aplicar la función a cada fila y agregar la nueva columna 'Condicion'
school_GP['extra'] = school_GP.apply(condiciones, axis=1)
```
